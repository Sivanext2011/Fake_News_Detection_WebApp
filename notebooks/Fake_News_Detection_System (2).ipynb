```python
import pandas as pd
import numpy as np
```

```python
true = pd.read_csv('True.csv')
fake = pd.read_csv('Fake.csv')
```

```python
true.head()
```

```python
fake.head()
```

```python
true['label'] = 1
fake['label'] = 0
```

```python
true.head()
```

```python
fake.head()
```

```python
# merging two datasets
news_data = pd.concat([fake, true], axis=0)
```

```python
news_data.head()
```

```python
news_data.tail()
```

```python
# Checking the null values
news_data.isnull().sum()
```

```python
# Dropping unnecessary columns
news_data = news_data.drop(['title', 'subject', 'date'], axis=1)
```

```python
news_data.head()
```

```python
# take entire data and reshuffale it
news_data = news_data.sample(frac=1)
```

```python
news_data.head()
```

```python
# reset indexing
news_data.reset_index(inplace=True)
```

```python
news_data.head()
```

```python
# drop unnecessary columns
news_data = news_data.drop(['index'], axis=1)
```

```python
news_data.head()
```

```python
import re
```

```python
def wordopt(text):
  # convert into lowercase
  text = text.lower()
  # remove URL
  text = re.sub('https?://\S+|www\.\S+', '', text)
  # remove HTML tags
  text = re.sub('<.*?>+', '', text)
  # remove punctuation
  text = re.sub(r'[^\w\s]', '', text)
  # remove digits
  text = re.sub('\w*\d\w*', '', text)
  # remove new line character
  text = re.sub('\n', '', text)
  return text
```

```python
# apply function over data
news_data['text'] = news_data['text'].apply(wordopt)
```

```python
news_data['text']
```

```python
# Spliting dataset into X and Y
X = news_data['text']
Y = news_data['label']
```

```python
X
```

```python
Y
```

```python
# divide into training and test date
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# convert textual data into numerical
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
XV_train = vectorization.fit_transform(X_train)
XV_test = vectorization.transform(X_test)
```

```python
XV_train
```

```python
XV_test
```

```python
# logistic regression model
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(XV_train, Y_train)
```

```python
# predict the model
pred_lr = LR.predict(XV_test)
```

```python
# score
LR.score(XV_test, Y_test)
```

```python
# print classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
DTC.fit(XV_train, Y_train)
```

```python
pred_dtc = DTC.predict(XV_test)
DTC.score(XV_test, Y_test)
```

```python
# print classification report test and prediction data
print(classification_report(Y_test, pred_dtc))
```

```python
# Random forest classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(XV_train, Y_train)
```

```python
predict_rfc = rfc.predict(XV_test)
rfc.score(XV_test, Y_test)
```

```python
print(classification_report(Y_test, predict_rfc))
```

```python
# gradient boosting classifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()
gbc.fit(XV_train, Y_train)
```

```python
pred_gbc = gbc.predict(XV_test)
gbc.score(XV_test, Y_test)
```

```python
print(classification_report(Y_test, pred_gbc))
```
